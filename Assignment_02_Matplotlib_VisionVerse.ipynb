## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
# Data
x = [-0.04, -0.02, 0.00, 0.02, 0.04]

# Create a figure object
fig = plt.figure()

# Add an axis to the figure canvas at [0,0,1,1]
ax = fig.add_axes([0, 0, 1, 1])

# Hide y-axis ticks and labels
ax.yaxis.set_visible(False)

# Set labels and titles
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

ax.plot()

# Show the plot
plt.show()
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# # Data
# x = [1, 2, 3, 4, 5]
# y = [2, 4, 6, 8, 10]

# Create a figure object
fig = plt.figure()

# Add two axes: ax1 and ax2
ax1 = fig.add_axes([0, 0, 1, 1])        # [left, bottom, width, height]
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])  # [left, bottom, width, height]

# Plot (x, y) on both axes
ax1.plot()
ax1.set_title('Ax1 Title')

ax2.plot()  # Red color for the plot on ax2
ax2.set_title('Ax2 Title')

# Show the plot
plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating data within the specified ranges
x = np.linspace(0, 100, 6)  # 0,20,40,60,80,100
y = np.linspace(50, 200, 4)  # 50,100,150,200
z = np.linspace(2000, 10000, 5)  # 2000,4000,6000,8000,10000

# Create two separate plots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot 1 - Left-hand side plot
ax1.plot(x, 2*x, linestyle='--', label='Main Plot')
# ax1.set_title('Main Plot')
ax1.set_xticks(x)  # Set x-axis ticks
ax1.set_yticks(y)  # Set y-axis ticks
ax1.legend()

# Plot 2 - Right-hand side plot
x = np.linspace(0, 100, 1000)  # Generate smoother x values
ax2.plot(x, x**2, color='r', label='Inserted Plot')
x = np.linspace(0, 100, 6)
ax2.set_xticks(x)  # Set x-axis ticks using specified x values
ax2.set_yticks(z)  # Set y-axis ticks
ax2.legend()

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```
